# 如何理解并使用Matplotlib

```python
#1 加载matplotlib包
import matplotlib.pyplot as plt
```

```python
#2 绘制空白，且不带坐标轴的图形区域

fig = plt.figure(figsize=(14,5))
```

```python
#3 绘制空白，带坐标轴的图形区域

fig, ax = plt.subplots(figsize=(14,5))
```

```python
#4 投射数据
fig, ax = plt.subplots(figsize=(14,5))
ax.plot([1,2,3,4],[1,4,2,3])
```

```python
#5 为坐标轴标记名称

fig, ax = plt.subplots(figsize=(14,5))
ax.plot([1,2,3,4],[1,4,2,3])
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
#6 编写图形标题
fig, ax = plt.subplots(figsize=(14,5))
ax.plot([1,2,3,4],[1,4,2,3])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Simple Plot')
```

```python
#7 为图形设置展示方式
fig, ax = plt.subplots(figsize=(14,5))  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], color='pink', marker='+', linestyle='dashed', markersize=12, linewidth=3);  # Plot some data on the axes.
ax.plot([1, 2, 3, 4], [2, 1, 3, 1], color='green', marker='+', linestyle='solid', markersize=12, linewidth=3);  # Plot some data on the axes.
ax.set_xlabel('x'); # Add an x-label to the axes.
ax.set_ylabel('y'); # Add a y-label to the axes.
ax.set_title('Simple Plot'); # Add a title to the axes.
```

```python
#8 添加图形中每条线的标签，并为图形添加图例；

fig, ax = plt.subplots(figsize=(14,5))  # Create a figure containing a single axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], color='pink', marker='+', linestyle='dashed', markersize=12, linewidth=3, label = 'Jan');  # Plot some data on the axes.
ax.plot([1, 2, 3, 4], [2, 1, 3, 1], color='green', marker='+', linestyle='solid', markersize=12, linewidth=3, label = 'Feb');  # Plot some data on the axes.
ax.plot([1, 2, 3, 4], [3, 2, 4, 5], color='blue', marker='*', linestyle='solid', markersize=12, linewidth=3, label = 'Mar');  # Plot some data on the axes.
ax.set_xlabel('x'); # Add an x-label to the axes.
ax.set_ylabel('y'); # Add a y-label to the axes.
ax.set_title('Simple Plot'); # Add a title to the axes.
ax.legend();
```
